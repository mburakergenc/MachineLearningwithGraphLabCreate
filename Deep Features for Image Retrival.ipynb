```python
import graphlab
```

# Load the CIFAR-10 Dataset

```python
image_train = graphlab.SFrame('image_train_data/')
image_test = graphlab.SFrame('image_test_data/')
```

# Train a nearest neighbors model for retrieving similar images

```python
knn_model = graphlab.nearest_neighbors.create(image_train, features=['deep_features'], label='id')
```

# Use image retrieval model to query for similar images

```python
graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]
cat['image'].show()
```

```python
knn_model.query(cat)
```

```python
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'], 'id')
```

```python
cat_neighbors = get_images_from_ids(knn_model.query(cat))
```

```python
cat_neighbors['image'].show()
```

```python
car = image_train[8:9]
car['image'].show()
```

```python
get_images_from_ids(knn_model.query(car))['image'].show()
```

# Lambda function to make things easier

```python
showNeighbors = lambda i: get_images_from_ids(knn_model.query(image_train[i:i+1]))['image'].show()
```

```python
showNeighbors(1250)
```

```python
showNeighbors(16)
```

```python
showNeighbors(210)
```

# Some flavors to play with

```python
image_train['label'].sketch_summary()
```

```python
cats = image_train[image_train['label'] == 'cat']
```

```python
dogs = image_train[image_train['label'] == 'dog']
```

```python
cars = image_train[image_train['label'] == 'automobile']
```

```python
birds = image_train[image_train['label'] == 'bird']
```

```python
cat_model = graphlab.nearest_neighbors.create(cats, features=['deep_features'], label='id')
```

```python
dog_model = graphlab.nearest_neighbors.create(dogs, features=['deep_features'], label='id')
```

```python
car_model = graphlab.nearest_neighbors.create(cars, features=['deep_features'], label='id')
```

```python
bird_model = graphlab.nearest_neighbors.create(birds, features=['deep_features'], label='id')
```

```python
cat_model.query(image_test[:1])
```

```python
dog_model.query(image_test[:1])
```

```python
 cat_model.query(image_test[:1])['distance'].mean()
```

```python
dog_model.query(image_test[:1])['distance'].mean()
```

```python
image_test_cat = image_test[image_test['label'] == 'cat']
```

```python
image_test_dog = image_test[image_test['label'] == 'dog']
```

```python
image_test_car = image_test[image_test['label'] == 'automobile']
```

```python
image_test_bird = image_test[image_test['label'] == 'bird']
```

```python
dog_cat_neighbors = cat_model.query(image_test_dog, k=1)
```

```python
dog_bird_neighbors = bird_model.query(image_test_dog, k=1)
```

```python
dog_automibile_neighbors = car_model.query(image_test_dog, k=1)
```

```python
dog_dog_neighbors = dog_model.query(image_test_dog, k=1)
```

```python
dogDistances = graphlab.SFrame({'dog-automobile': dog_automibile_neighbors['distance'], 'dog-bird': dog_bird_neighbors['distance'], 'dog-cat': dog_cat_neighbors['distance'], 'dog-dog':dog_dog_neighbors['distance']})
```
